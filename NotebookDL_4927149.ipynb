```python
import pandas as pd
import numpy as np
import tensorflow as tf
import optuna
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam


SEED = 1234

# Cargar el conjunto de datos
datos = pd.read_csv("Temp_Asu20092021.csv")

# Convertir la columna 'Fecha' al tipo datetime y establecerla como índice
datos['Fecha'] = pd.to_datetime(datos['Fecha'])
datos.set_index('Fecha', inplace=True)

# Filtrar datos a partir de 2019
datos = datos[datos.index.year >= 2019]

# Interpolar valores faltantes y resamplear a frecuencia horaria
datos = datos.resample('H').interpolate(method='linear')

# Calcular percentiles diarios (95, 50, 5)
percentiles_diarios = datos.resample('D')['Temperatura'].agg(
    Percentil_95=lambda x: np.percentile(x, 95),
    Mediana=lambda x: np.percentile(x, 50),
    Percentil_5=lambda x: np.percentile(x, 5)
)

# Crear variables objetivo para el día siguiente
percentiles_diarios['Percentil_95_siguiente'] = percentiles_diarios['Percentil_95'].shift(-1)
percentiles_diarios['Mediana_siguiente'] = percentiles_diarios['Mediana'].shift(-1)
percentiles_diarios['Percentil_5_siguiente'] = percentiles_diarios['Percentil_5'].shift(-1)

# Eliminar filas con valores faltantes después del desplazamiento
percentiles_diarios.dropna(inplace=True)

# Separar en variables de entrada y salida
X = percentiles_diarios[['Percentil_95', 'Mediana', 'Percentil_5']].values
y = percentiles_diarios[['Percentil_95_siguiente', 'Mediana_siguiente', 'Percentil_5_siguiente']].values

# Escalar características
scaler_X = StandardScaler()
X_escalado = scaler_X.fit_transform(X)

# Separar en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X_escalado, y, test_size=0.2, random_state=SEED, shuffle=False
)

# Reshape para LSTM
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# 2. Definir función objetivo para Optuna
def objetivo(trial):
    unidades_lstm = trial.suggest_int('unidades_lstm', 10, 100)
    lr = trial.suggest_loguniform('learning_rate', 1e-5, 1e-2)
    modelo = Sequential([
    # Primera capa LSTM (establece return_sequences=True para permitir más capas LSTM después)
    LSTM(units=mejores_params["unidades_lstm"], input_shape=(1, X_train.shape[2]), return_sequences=True),
    # Segunda capa LSTM (solo se requiere return_sequences=True si se planea añadir otra LSTM)
    LSTM(units=mejores_params["unidades_lstm"] // 2, return_sequences=True),
    # Capa LSTM final sin return_sequences
    LSTM(units=mejores_params["unidades_lstm"] // 4),
    # Añadir una capa densa intermedia para una mejor combinación de características
    Dense(32, activation='relu'),
    # Capa de salida
    Dense(3)  # Tres salidas para Percentil 95, Mediana y Percentil 5
])


    modelo.compile(optimizer=Adam(learning_rate=lr), loss='mse', metrics=['mae'])

    # Entrenar el modelo
    modelo.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0, shuffle=False)

    predicciones = modelo.predict(X_test)
    mae = mean_absolute_error(y_test, predicciones)
    return mae

# 3. Optimización con Optuna
estudio = optuna.create_study(direction='minimize')
estudio.optimize(objetivo, n_trials=40)

# 4. Imprimir los mejores hiperparámetros
mejores_params = estudio.best_params
print(f'Mejor número de unidades LSTM: {mejores_params["unidades_lstm"]}')
print(f'Mejor tasa de aprendizaje: {mejores_params["learning_rate"]}')

# 5. Entrenar el modelo final con los mejores hiperparámetros
modelo_final = Sequential([
    # Primera capa LSTM (establece return_sequences=True para permitir más capas LSTM después)
    LSTM(units=mejores_params["unidades_lstm"], input_shape=(1, X_train.shape[2]), return_sequences=True),
    # Segunda capa LSTM
    LSTM(units=mejores_params["unidades_lstm"] // 2, return_sequences=True),
    # Capa LSTM final sin return_sequences
    LSTM(units=mejores_params["unidades_lstm"] // 4),
    # Capa densa intermedia para una mejor combinación de características
    Dense(32, activation='relu'),
    # Capa de salida
    Dense(3)  # Tres salidas para Percentil 95, Mediana y Percentil 5
])

modelo_final.compile(optimizer=Adam(learning_rate=mejores_params["learning_rate"]), loss='mse', metrics=['mae'])

# Historial de entrenamiento
historial = modelo_final.fit(
    X_train, y_train, epochs=10, batch_size=32,
    validation_data=(X_test, y_test), verbose=1, shuffle=False
)

# 6. Evaluar el modelo final
mse, mae = modelo_final.evaluate(X_test, y_test, verbose=0)
print(f'MAE en test: {mae:.2f}')
print(f'MSE en test: {mse:.2f}')

# 7. Predicciones y gráficos de resultados
predicciones_finales = modelo_final.predict(X_test)

# Crear gráficos de dispersión individuales para cada valor predicho (Percentil 95, Mediana, Percentil 5)
percentiles = ['Percentil 95', 'Mediana', 'Percentil 5']
for i in range(3):
    plt.figure(figsize=(6, 6))
    plt.scatter(y_test[:, i], predicciones_finales[:, i], alpha=0.5, label=f'Predicción {percentiles[i]}')
    plt.plot([y_test[:, i].min(), y_test[:, i].max()],
             [y_test[:, i].min(), y_test[:, i].max()], 'k--', color='red', label="Línea 45°")
    plt.xlabel(f'Valores reales - {percentiles[i]}')
    plt.ylabel(f'Predicciones - {percentiles[i]}')
    plt.title(f'Dispersión entre Valores Reales y Predicciones ({percentiles[i]})')
    plt.legend()
    plt.show()

# Curva de aprendizaje
plt.figure(figsize=(10, 5))
plt.plot(historial.history['loss'], label='Pérdida de entrenamiento')
plt.plot(historial.history['val_loss'], label='Pérdida de validación')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.title('Curva de Aprendizaje')
plt.legend()
plt.show()


# 10 Mejores resultados de Optuna
top_trials = sorted(estudio.trials, key=lambda x: x.value)[:10]
print("Top 10 Mejores Resultados de Optuna:")
for i, trial in enumerate(top_trials):
    print(f"Trial {i+1} - MAE: {trial.value}, Parámetros: {trial.params}")
```
